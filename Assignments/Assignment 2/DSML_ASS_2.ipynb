```python
# decribing varience of each feature.

df = pd.DataFrame(data)
print(df.var())
```

```python
# finding range for each feature

for dt in data :
  if(dt != "Genre") :
    range = data[dt].max() - data[dt].min()
    print(dt , " = ", range)
```

b. Data Visualization-Create a histogram for each feature in the dataset to illustrate the feature
distributions

```python
# histogram for age
plt.hist(data["Age"])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title('histogram')
plt.show()

# histogram for Annual income
plt.hist(data.iloc[:,3])
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.title('histogram')
plt.show()

# histogram for spending score :
plt.hist(data.iloc[:,4])
plt.xlabel("Spending Score")
plt.ylabel("Frequency")
plt.title('histogram')
plt.show()
```
